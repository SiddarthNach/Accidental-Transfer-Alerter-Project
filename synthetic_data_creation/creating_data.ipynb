```python
from faker import Faker
import numpy as np
import random
```

```python
faker = Faker()
random.seed(42)
np.random.seed(42)
```

```python
from datetime import timedelta, datetime
```

```python
#lets simulate data from the beginning of this year
start_date = datetime(2025, 1, 1) 
```

```python
bank_data['timestamp'] = bank_data['step'].apply(lambda x: start_date + timedelta(minutes=int(x * 10)))
```

```python
#now lets assign a city location to each of these transactions
cities = ['New York', 'Chicago', 'Los Angeles', 'Houston', 'Phoenix', 'Seattle', 'Boston']
user_city_map = {}

bank_data['location'] = bank_data['nameOrig'].apply(lambda user: user_city_map.setdefault(user, random.choice(cities)))
```

```python
#generate a random account number and phone
bank_data['receiverAccount'] = ['ACCT' + ''.join(random.choices('0123456789', k=8)) for _ in range(len(bank_data))]
bank_data['receiverPhone'] = ['+1' + ''.join(random.choices('0123456789', k=10)) for _ in range(len(bank_data))]
```

```python
bank_data.head(10)
```

```python
bank_data.head(10)  
```

```python
bank_data['isAccident'] = 0
```

```python
#calculating if the amount transfered is over user mean
user_mean = bank_data.groupby('nameOrig')['amount'].transform('mean')
bank_data.loc[bank_data['amount'] > 5 * user_mean, 'isAccident'] = 1
```

```python
location_anomaly_indices = np.random.choice(bank_data.index, size=int(len(bank_data) * 0.03), replace=False)
bank_data.loc[location_anomaly_indices, 'location'] = bank_data.loc[location_anomaly_indices, 'location'].apply(
    lambda loc: random.choice([c for c in cities if c != loc])
)
bank_data.loc[location_anomaly_indices, 'isAccident'] = 1
```

```python
#assigning the users a consistent phone address
user_phone_map = {user: '+1' + ''.join(random.choices('0123456789', k=10)) for user in bank_data['nameOrig'].unique()}
bank_data['expectedPhone'] = bank_data['nameOrig'].map(user_phone_map)
```

```python
#randomly changing a few phone numbers to simulate accidental transfers
phone_anomaly_indices = np.random.choice(bank_data.index, size=int(len(bank_data) * 0.03), replace=False)
bank_data.loc[phone_anomaly_indices, 'receiverPhone'] = ['+1' + ''.join(random.choices('0123456789', k=10)) for _ in range(len(phone_anomaly_indices))]
```

```python
#labelling for mismatch in usual phone numbers and the receiver phone number
bank_data.loc[bank_data['receiverPhone'] != bank_data['expectedPhone'], 'isAccident'] = 1
```

```python
import re

# Normalize phone number to consistent format
def normalize_phone(phone):
    digits = re.sub(r'\D', '', phone)
    return '+1' + digits[-10:]  # last 10 digits

# Assign consistent phone per user
user_phone_map = {}
receiver_phones = []

for user in bank_data['nameOrig']:
    if user not in user_phone_map:
        user_phone_map[user] = '+1' + ''.join(random.choices('0123456789', k=10))
    receiver_phones.append(user_phone_map[user])

bank_data['receiverPhone'] = receiver_phones

# Inject anomalies into 3%
phone_anomaly_indices = np.random.choice(bank_data.index, size=int(len(bank_data) * 0.03), replace=False)
bank_data.loc[phone_anomaly_indices, 'receiverPhone'] = ['+1' + ''.join(random.choices('0123456789', k=10)) for _ in range(len(phone_anomaly_indices))]

# Map expected phones
bank_data['expectedPhone'] = bank_data['nameOrig'].map(user_phone_map)

# Normalize both columns
bank_data['receiverPhone'] = bank_data['receiverPhone'].apply(normalize_phone)
bank_data['expectedPhone'] = bank_data['expectedPhone'].apply(normalize_phone)

# Flag accidents where phones differ
bank_data['isAccident'] |= bank_data['receiverPhone'] != bank_data['expectedPhone']

# Cleanup
bank_data.drop(columns='expectedPhone', inplace=True)
```
